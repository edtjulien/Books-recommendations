```python
df_meta = pd.read_json('../output/books-meta-data.json', lines=True)
```

```python
import ast

def is_keep_tag(tag):
    return tag in list(THEMES.keys())

def tags_to_cols(df, col_name):
    df1 = df.copy()
    for index,row in df.iterrows():
        tags_as_string = row[col_name]
        tags = ast.literal_eval(tags_as_string)

        for tag in tags:
            if not is_keep_tag(tag[0].strip()):
                continue
            tag_name = 'tag_'+tag[0].strip().replace(' ','_').lower()
            df1.loc[df1.index == index, tag_name] = tag[1]

    return df1
```

```python
df2 = tags_to_cols(df_meta, 'tags')
df2
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity

df_tags = df2.filter(regex='^tag_',axis=1)
df_tags = df_tags.fillna(0)
tags = MinMaxScaler().fit_transform(df_tags)
tags_score = cosine_similarity(tags)
```
