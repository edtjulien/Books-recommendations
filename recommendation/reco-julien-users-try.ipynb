```python
from tqdm import tqdm

df_pearson = pd.DataFrame(index=pivot.index,columns=pivot.index)
np_pearson = df_pearson.to_numpy()

pivot = pivot.copy()
pivot2 = pivot.copy()
pearson_done = []
i = 0
for row_id, row_values in tqdm(pivot.iterrows()):
    j = -1
    for col_id, col_values in pivot2.iterrows():
        j += 1
        if str(row_id)+' '+str(col_id) in pearson_done:
            continue

        pearson_value = pearsonr(row_values.to_numpy().reshape(-1),col_values.to_numpy().reshape(-1)).statistic
        # df_pearson.loc[df_pearson.index == row_id, col_id] = pearson_value
        # df_pearson.loc[df_pearson.index == col_id, row_id] = pearson_value
        np_pearson[i,j] = pearson_value
        np_pearson[j,i] = pearson_value

        pearson_done.append(str(row_id)+' '+str(col_id))
        pearson_done.append(str(col_id)+' '+str(row_id))
    if i > 0:
        break
    i += 1

df_pearson = pd.DataFrame(np_pearson,index=pivot.index,columns=pivot.index)
```
